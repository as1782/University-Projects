```python
#Part A

def MinDist(input):
    
    #Input list is sorted in acending order and the absolute diffrence between second and first element is found
    
    input.sort()
    
    min_dist = abs(input[1] - input[0])
    
    #For loop iterates the rest of the elements in the input list from the 3rd element, the absolute distance is measured
    
    for i in range(2, len(input)):
        
        dist = abs(input[i] - input[i-1])
    
    #If calculated distance is smaller the minmum distance, the variable is updated to new minimum value
    
        if dist < min_dist:
            
            min_dist = dist
               
    return min_dist

print ("Output:", MinDist([8,10,14,20,31,41])) 
```

```python
#Part B

def MaxMinDist(input):
    
    #The function takes the max min distance for each elements in the input list
    
    return max([MinDist(List) for List in input])

print ("Output:", MaxMinDist([[5,9,1,3],[3,4,1,1]]))
```

```python
#Part C

print("Output:", MaxMinDist([[5,2,1,6],[10,0,4],[9,18,1],[100,100,27,9,18],[28,30]]))
```

```python
#Part D

np.random.seed(99)

#Generates an array of random numbers

arr = np.random.randint(0, 3001, size=1000)

#Reshapes the array into 50x20 matrix

arr_2d = arr.reshape(50, 20)

print ("Output:", MaxMinDist(arr_2d))
```

<h2>Problem 4<h2> 

```python
#Create an array for x values

x = np.arange(-7, 7, 0.5)

#Set y values

y1 = np.sin(x) / 3

y2 = -np.cos(x)

#Create the figure and the subplots

fig,ax = plt.subplots(figsize=(15,8))

ax.set_ylim([-1.6, 1.6])

ax.set_xticks(np.arange(-7, 8, 1.56))

ax.set_yticks(np.arange(-1.6, 1.7, 0.8))

ax.plot(x, y1, color='hotpink', marker='s', linestyle='dashed', label='f(x)=1/3Sin(x)')

ax.plot(x, y2, color='C0', marker='|', mec='C0', mfc='C0', label='f(x)=-Cos(x)')

#Create the titles, legends and labels 

ax.set_title("Plot of Sin and Cos Graphs", fontsize = 15)

ax.set_xlabel("X", fontsize=15)

ax.set_ylabel("f(x)", fontsize=15)

ax.legend()

plt.show
```

```python
#Part A

df = pd.read_csv('harrypotter_dataset.csv')

print(df)

print("Number of rows:", len(df))
```

```python
#Part B 

print("Column Headings:", df.columns)
```

```python
#Part C

#Remove any white space from the strings of column headings

df.columns = df.columns.str.strip()

print("Column Headings:", df.columns)
```

```python
#Part D 

df['Runtime (in hours)'] = round(df['Runtime (in minutes)'] / 60, 1)
```

```python
#Part E

df['UK_US_Same_Date'] = df['UK Movie release date'] == df['US Movie release date']
```

```python
#Part F

total_chapters = df['No. of chapters'].sum()

book7_chapters = df[df['Book index'] == 7]['No. of chapters'].sum()

probability_book7 = book7_chapters / total_chapters

print("Probability of a chapter being from Book 7:", probability_book7)

movies_1_3_5_7 = df.loc[df['Movie index'].isin([1, 3, 5, 7])]

total_runtime_of_1_3_5_7 = movies_1_3_5_7['Runtime (in minutes)'].sum()

total_runtime = df['Runtime (in minutes)'].sum()

probability_movies_1_3_5_7 = total_runtime_of_1_3_5_7 / total_runtime

print("Probability of choosing a minute from movies 1, 3, 5, or 7:", probability_movies_1_3_5_7)

percent_of_same_date = df['UK_US_Same_Date'].mean() * 100

print("Percentage of movies released on same date in UK and US:", percent_of_same_date, "%")
```

```python
#Part G 

df_nineties = df[df['Book release year'] < 2000]
```

```python
#Part H

y = df['Box office gross'].str.replace('$', '', regex=True)

y = y.astype(float)

fig, ax = plt.subplots(figsize=(15,8))

ax.plot(y, color='C1', marker='d')

ax.set_xticks(np.arange(1, 9))

ax.set_title("Box Office Grossing per Harry Potter Movie", fontsize=18)

ax.set_xlabel("Movie Index", fontsize=15)

ax.set_ylabel("Box Office Gross", fontsize=15)

plt.show()
```

<h2>Problem 6<h2>

```python
fig, ax = plt.subplots(figsize=(15, 8))

ax.scatter(df['Runtime (in hours)'], df['Box office gross'], color='C1', marker='x')

ax.set_xlabel('Runtime (in hours)', fontsize=15)

ax.set_ylabel('Box office gross', fontsize=15)

ax.set_title('Scatterplot of Runtime (in hours) vs Box office gross', fontsize=18)

plt.show()

corr_coefficient = round(df['Runtime (in hours)'].corr(df['Box office gross']), 2)

print("Correlation coefficient between 'Runtime (in hours)' and 'Box office gross':", corr_coefficient)
```

My initial hypothesis was that longer movies could mean higher Box Office Gross values as people may feel they got thier money's worth as a result increasing the demand for a movie ticket. After visualising the data, additionally finding the correlation coefficient of (-0.61). This indicates a mildly negative correlation between the 2 variables. This could be because, longer movies may cause fatigue which may put off people from paying for the ticket. Interpretation must be done with caution as the data does not indicate causality.
